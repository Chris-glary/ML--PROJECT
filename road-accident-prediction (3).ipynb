```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/road-accident-casualties-dataset/accident data.csv")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
import matplotlib.pyplot as plt
df.hist()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.scatterplot(data=df,y= 'Number_of_Casualties',x='Number_of_Vehicles',hue='Accident_Severity',ax=axes[0,0])
axes[0, 0].set_title('Accident Severity')

sns.scatterplot(data=df,y= 'Number_of_Casualties',x='Number_of_Vehicles',hue='Road_Type',ax=axes[0,1])
axes[0, 1].set_title('Road Type')

sns.scatterplot(data=df,y= 'Number_of_Casualties',x='Number_of_Vehicles',hue='Urban_or_Rural_Area',ax=axes[1,0])
axes[1, 0].set_title('Urban or Rural Area')

sns.scatterplot(data=df,y= 'Number_of_Casualties',x='Number_of_Vehicles',hue='Light_Conditions',ax=axes[1,1])
axes[1, 1].set_title('Light Conditions')

plt.tight_layout()
plt.show()
```
